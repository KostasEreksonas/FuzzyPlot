```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
x_asis = np.arange(0, 12, 1)
labai_blogai = [-1000, 2, 2, 4]
blogai = [2, 4, 4, 6]
patenkinamai = [4, 6, 6, 8]
gerai = [6, 8, 8, 10]
puikiai = [8, 10, 10, 1000]

x1 = fuzz.trapmf(x_asis, labai_blogai)
x2 = fuzz.trapmf(x_asis, blogai)
x3 = fuzz.trapmf(x_asis, patenkinamai)
x4 = fuzz.trapmf(x_asis, gerai)
x5 = fuzz.trapmf(x_asis, puikiai)

fig, ax = plt.subplots(figsize=(10,10))

plt.xticks(np.arange(min(x_asis), max(x_asis)+1, 1))
plt.yticks(np.arange(0, 2, 0.1))

ax.vlines(x = 7, ymin = 0, ymax = 0.5, color = 'violet', linewidth = 3.0, linestyle = 'dashed')
ax.hlines(y = 0.5, xmin = 0, xmax = 7, color = 'violet', linewidth = 3.0, linestyle = 'dashed')

ax.plot(x_asis, x1, 'black', linewidth=1.5, label='Labai blogai')
ax.plot(x_asis, x2, 'red', linewidth=1.5, label='Blogai')
ax.plot(x_asis, x3, 'orange', linewidth=1.5, label='Patenkinamai')
ax.plot(x_asis, x4, 'yellow', linewidth=1.5, label='Gerai')
ax.plot(x_asis, x5, 'blue', linewidth=1.5, label='Puikiai')

ax.grid()
ax.set_title('ini vertinimas')  
ax.legend();
```
